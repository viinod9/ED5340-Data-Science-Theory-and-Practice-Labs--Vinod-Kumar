## 1. Write a program that takes coefficients A, B, C, D, and E as inputs representing a 4th degree polynomial in the form Ax^4 + Bx^3 + Cx^2 + Dx + E. Calculate the values of this polynomial for x in the range from -100 to 100, with constant discrete intervals.

Store the resulting x and y values as a NumPy array, where x represents the input values, and y represents the corresponding output values of the polynomial. Finally, use Matplotlib to plot the graph using the generated NumPy array.

```python
import numpy as np
import matplotlib.pyplot as plt

def polynomial(A, B, C, D, E, x):
  return A*x**4 + B* x**3 + C* x**2 + D* x + E

A = int(input("Enter the value of A : \n"))
B = int(input("Enter the value of B : \n"))
C = int(input("Enter the value of C : \n"))
D = int(input("Enter the value of D : \n"))
E = int(input("Enter the value of E : \n"))

print(f'The graph of Function f(x) = {A}x^4 + {B} x^3 + {C} x^2 + {D} x + {E} is ')
x_val = np.linspace(-100, 100, 100)
y_val = polynomial(A, B,C,D,E, x_val)
plt.plot(x_val, y_val)
plt.xlabel("x")
plt.ylabel("y")
plt.title(f'f(x) = {A}x^4 + {B} x^3 + {C} x^2 + {D} x + {E}')
plt.show()
```

##(2). Suppose you have a dictionary containing information about monthly sales for different products over a period of time. The dictionary has the following structure.

sales_data = {

    'Product': ['A', 'B', 'A', 'C', 'B', 'C', 'A', 'B', 'C'],

    'Month': ['Jan', 'Jan', 'Feb', 'Feb', 'Mar', 'Mar', 'Apr', 'Apr', 'Apr'],

    'Sales': [100, 150, 200, 120, 180, 220, 90, 110, 130]

}

Write a Python script to convert this dictionary into a pandas DataFrame, calculate the total sales for each product over the entire period, and then create a bar plot using matplotlib to visualize the total sales for each product.-

```python
import pandas as pd

sales_data = {'Product': ['A', 'B', 'A', 'C', 'B', 'C', 'A', 'B', 'C'],
              'Month': ['Jan', 'Jan', 'Feb', 'Feb', 'Mar', 'Mar', 'Apr', 'Apr', 'Apr'],
              'Sales': [100, 150, 200, 120, 180, 220, 90, 110, 130]}

data = pd.DataFrame(sales_data)
data
```

```python
total_sales = data.groupby('Product')['Sales'].sum()
print(total_sales)
total_sales.plot.bar()
plt.xlabel('Product')
plt.ylabel('Total Sales')
plt.title('Total Sales for Each Product')
plt.show()
```

##(4). For the function J(w) = w^2 + (54/w), implement the bracketing method
(choose your own a, b, n).

```python
def J(w):
    return w**2 + 54/w

def exhaustive_search(a, b, num_points):
    min_value = float('inf')
    min_point = None

    # Generate points within the interval [a, b]
    for w in np.linspace(a, b, num_points):
        current_value = J(w)
        if current_value < min_value:
            min_value = current_value
            min_point = w

    return min_point, min_value

a = 1
b = 10
num_points = 10
min_point, min_value = exhaustive_search(a, b, num_points)

print("Minimum point:", min_point)
print("Minimum value:", min_value)
```
