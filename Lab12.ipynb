```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, confusion_matrix
```

# Iris DataSet

```python
# Load Iris dataset
iris = datasets.load_iris()
print('First 5 rows of iris dataset : \n',iris.data[:5])
print('Size of Iris Data Set : \n',iris.data.shape)
```

```python
# We'll use only the first two features (sepal length and width)
X_iris = iris.data[:, :2]
y_iris = iris.target
```

```python
# Visualize Iris dataset
plt.figure(figsize=(8, 6))
plt.scatter(X_iris[:, 0], X_iris[:, 1], c=y_iris, cmap=plt.cm.Set1, edgecolor='k')
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.title('Iris dataset')
plt.show()
```

```python
# Splitting the Iris dataset into training and testing sets
X_train_iris, X_test_iris, y_train_iris, y_test_iris = train_test_split(X_iris, y_iris, test_size=0.3, random_state=42)
```

```python
# Initialize and train SVM model for Iris dataset
svm_iris = SVC(kernel='linear', C=0.1, random_state=42)
svm_iris.fit(X_train_iris, y_train_iris)
```

```python
# Testing the model on Iris dataset
y_pred_iris = svm_iris.predict(X_test_iris)
print('Testing Results : \n',y_test_iris)
print('Predicted Results : \n',y_pred_iris)
accuracy_iris = accuracy_score(y_test_iris, y_pred_iris)
print("Accuracy on Iris dataset : \n", accuracy_iris)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets

# Load the Iris dataset
iris = datasets.load_iris()
X = iris.data
y = iris.target

# Standardize the data
X_centered = (X - np.mean(X, axis=0))/np.std(X,axis=0)

# Calculate covariance matrix
cov_matrix = np.cov(X_centered, rowvar=False)
print('Covariance Matrix : \n',cov_matrix)

# Perform eigen decomposition
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)

# Sort eigenvalues and eigenvectors
sorted_indices = np.argsort(eigenvalues)[::-1]
eigenvalues = eigenvalues[sorted_indices]
eigenvectors = eigenvectors[:, sorted_indices]

# Project data onto the first three principal components
projected_data = X_centered.dot(eigenvectors[:, :3])
print('Data after standardization and Eigen Decomposition : \n',projected_data[:5])

# Plotting the results
fig = plt.figure(figsize=(4, 3))
plt.clf()

ax = fig.add_subplot(111, projection="3d", elev=48, azim=100)
ax.set_position([0, 0, 0.95, 1])

plt.cla()

for name, label in [("Setosa", 0), ("Versicolour", 1), ("Virginica", 2)]:
    ax.text3D(
        projected_data[y == label, 0].mean(),
        projected_data[y == label, 1].mean() + 1.5,
        projected_data[y == label, 2].mean(),
        name,
        horizontalalignment="center",
        bbox=dict(alpha=0.5, edgecolor="w", facecolor="w"),
    )

# Reorder the labels to have colors matching the cluster results
y = np.choose(y, [1, 2, 0]).astype(float)

ax.scatter(projected_data[:, 0], projected_data[:, 1], projected_data[:, 2], c=y, cmap=plt.cm.nipy_spectral, edgecolor="k")

ax.xaxis.set_ticklabels([])
ax.yaxis.set_ticklabels([])
ax.zaxis.set_ticklabels([])

plt.show()
```

```python
from sklearn import datasets, decomposition

np.random.seed(5)

iris = datasets.load_iris()
X = iris.data
y = iris.target

# Create a figure for the plot
fig = plt.figure(1, figsize=(4, 3))
plt.clf()

# Add a subplot with 3D projection
ax = fig.add_subplot(111, projection="3d", elev=48, azim=100)
ax.set_position([0, 0, 0.95, 1])

# Clear the current plot
plt.cla()

# Perform PCA with 3 components
pca = decomposition.PCA(n_components=3)
pca.fit(X)
X = pca.transform(X)
print('Data after pca using scikit learn : \n',X[:5])

# Plot the data points with labels as text
for name, label in [("Setosa", 0), ("Versicolour", 1), ("Virginica", 2)]:
    ax.text3D(
        X[y == label, 0].mean(),
        X[y == label, 1].mean() + 1.5,
        X[y == label, 2].mean(),
        name,
        horizontalalignment="center",
        bbox=dict(alpha=0.5, edgecolor="w", facecolor="w"),
    )

# Reorder the labels to have colors matching the cluster results
y = np.choose(y, [1, 2, 0]).astype(float)

# Scatter plot the transformed data points
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=y, cmap=plt.cm.nipy_spectral, edgecolor="k")

# Set tick labels for each axis
ax.xaxis.set_ticklabels([])
ax.yaxis.set_ticklabels([])
ax.zaxis.set_ticklabels([])

# Show the plot
plt.show()
```
