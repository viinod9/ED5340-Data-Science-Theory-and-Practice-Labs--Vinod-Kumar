##(5) WAP to plot a 3-d graph of the helical wave signal using the scatter method and normal line method. Plot them separately and specify legend.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define helix parameters
radius = 1
theta = np.linspace(0, 10 * np.pi, 1000)

# Generate angles for the helix

# Calculate x, y, z coordinates for the helix
x = radius * np.cos(theta)
y = radius * np.sin(theta)
z = theta

# Scatter plot
fig = plt.figure(figsize=(8, 6))

# Create the first subplot for scatter method
ax1 = fig.add_subplot(211, projection='3d')
ax1.scatter(x, y, z)
ax1.set_title('Helical Wave Signal (Scatter Method)')
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_zlabel('Z')
ax1.legend(['Helical Wave'])

# Create the second subplot for normal line method
ax2 = fig.add_subplot(212, projection='3d')
ax2.plot(x, y, z)
ax2.set_title('Helical Wave Signal (Normal Line Method)')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_zlabel('Z')
ax2.legend(['Helical Wave'])

plt.show()
```

##(8) .Take a N X M integer array matrix with space separated elements ( N = rows and M  = columns). Your task is to print the transpose and flatten results using numpy

```python
def transflatt(matrix):
    # Convert input list to numpy array
    mat_arr = np.array(matrix).reshape((N, M))

    # Transpose the matrix
    trans_mat = np.transpose(mat_arr)

    # Flatten the matrix
    flat_mat = mat_arr.flatten()
    print("Matrix : ")
    print(mat_arr)
    print("\nTranspose : ")
    print(trans_mat)
    print("\nFlatten : ")
    print(flat_mat)

N = int(input("Enter number of rows N: "))
M = int(input("Enter number of columns M: "))
matrix = []
print("Enter matrix elements:")
for i in range(N):
    for j in range(M):
        p = int(input(f'{i+1}{j+1} index element of matrix: '))
        matrix.append(p)
transflatt(matrix)
```

##(10). Use the central difference method to find the first and second order derivatives of the function. Use the following function for testing the result.

f(x) = 3x**2 + 2x

```python
def f(x):
    return 3 * x**2 + 2 * x

def first_der(f, x, h):
    return (f(x + h) - f(x - h)) / (2 * h)

def second_der(f, x, h):
    return (f(x + h) - 2 * f(x) + f(x - h)) / (h**2)

x = 2
h = 0.01

first_der = first_der(f, x, h)
second_der = second_der(f, x, h)

print("First derivative at x =", x, "is:", first_der)
print("Second derivative at x =", x, "is:", second_der)
```
